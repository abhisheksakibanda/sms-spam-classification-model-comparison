```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import nltk
import string
from gensim.test.utils import common_texts, get_tmpfile
from gensim.models import Word2Vec
from collections import Counter
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
# Read SMS Dataset
sms_spam = pd.read_csv('spam.csv', encoding='latin-1', usecols=[0, 1])
sms_spam.rename(columns={'v1': 'label', 'v2': 'text'}, inplace=True)
sms_spam.head()
```

```python
# Text preprocessing
corpus = []
for i in range(0, len(sms_spam)):
    # Remove non-alphabetic characters, convert to lowercase, and split into words
    review = re.sub('[^a-zA-Z]', ' ', sms_spam['text'][i])
    review = review.lower()
    review = review.split()

    # Apply stemming and remove stopwords
    ps = PorterStemmer()
    nltk.download('stopwords')
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
# Create a CountVectorizer and transform the corpus into a bag-of-words representation
cv = CountVectorizer()
X = cv.fit_transform(corpus).toarray()
y = sms_spam.iloc[:, 0].values
```

```python
# Display some information about the processed data
print(cv.get_feature_names_out()[0:20])
print(cv.get_feature_names_out()[-20:])
```

```python
# Features Count
print(X.shape)
print(y.shape)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
# Extract words from spam and ham messages
ham_words = ''
spam_words = ''
spam = sms_spam[sms_spam.label == 'spam']
ham = sms_spam[sms_spam.label == 'ham']
```

```python
# Process spam messages
for val in spam.text:
    text = re.sub('[^a-zA-Z]', ' ', val)
    text = text.lower()
    text = text.split()
    ps = PorterStemmer()
    text = [ps.stem(word) for word in text if not word in set(stopwords.words('english'))]
    for words in text:
        spam_words = spam_words + words + ' '

# Process ham messages
for val in ham.text:
    text = re.sub('[^a-zA-Z]', ' ', val)
    text = text.lower()
    text = text.split()
    ps = PorterStemmer()
    text = [ps.stem(word) for word in text if not word in set(stopwords.words('english'))]
    for words in text:
          ham_words = ham_words + words + ' '
```

```python
# Define a function for text processing
def text_process(text):
    """
    Takes in a string of text, then performs the following:
    1. Remove all punctuation
    2. Remove all stopwords
    3. Returns a list of the cleaned text
    """
    STOPWORDS = stopwords.words('english') + ['u', 'ü', 'ur', '4', '2', 'im', 'dont', 'doin', 'ure']
    # Check characters to see if they are in punctuation
    nopunc = [char for char in text if char not in string.punctuation]

    # Join the characters again to form the string.
    nopunc = ''.join(nopunc)
    
    # Now just remove any stopwords
    return ' '.join([word for word in nopunc.split() if word.lower() not in STOPWORDS])
```

```python
# Apply text processing to create a new 'clean_text' column in the dataset
sms_spam['clean_text'] = sms_spam.text.apply(text_process)
sms_spam.head()
sms_spam['clean_text'].fillna("unknown", inplace=True)
```

```python
# Analyze the most common words in ham messages
words = sms_spam[sms_spam.label=='ham'].clean_text.apply(lambda x: [word.lower() for word in x.split()])
ham_words = Counter()

for msg in words:
    ham_words.update(msg)
    
ham_words.most_common(50)
```

```python
# Analyze the most common words in spam messages
words = sms_spam[sms_spam.label=='spam'].clean_text.apply(lambda x: [word.lower() for word in x.split()])
spam_words = Counter()

for msg in words:
    spam_words.update(msg)
    
spam_words.most_common(50)
```

```python
X = sms_spam['clean_text']
Y = sms_spam['label']
```

```python
# Vectorize the text data using CountVectorizer with specific settings
vect = CountVectorizer(min_df=5, ngram_range=(1,2)).fit(X)

X_vec = vect.transform(X)
# Display information about the vectorized data
len(vect.get_feature_names_out())
```

```python
# Split the vectorized data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_vec, y, test_size=0.2, random_state = 0)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Logistic Regression
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.linear_model import LogisticRegression

# Create a pipeline for the Logistic Regression model
model_lr = Pipeline([('tfidf', TfidfTransformer()),
                   ('model', LogisticRegression()),
                   ])

# Train the Logistic Regression model
model_lr.fit(X_train, y_train)

# Predict on the test set
ytest = np.array(y_test)
pred_y = model_lr.predict(X_test)
```

```python
# Plot Confusion Matrix
conf_matrix_lr = confusion_matrix(y_test, pred_y)
print(conf_matrix_lr)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_lr, annot=True, fmt='d', cmap='Blues')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
print(conf_matrix_lr)
```

```python
# Evaluate Logistic Regression model performance
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

print('accuracy: %s' %accuracy_score(pred_y, y_test))
print(classification_report(ytest, pred_y))
```

```python
# Multinomial Naive Bayes
from sklearn.naive_bayes import MultinomialNB

# Create pipeline
model_nb = Pipeline([('tfidf', TfidfTransformer()),
                   ('model', MultinomialNB()),
                   ])

#Train Model
model_nb.fit(X_train, y_train)

# Predict on test set
ytest = np.array(y_test)
pred = model_nb.predict(X_test)
```
